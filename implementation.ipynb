```python
from PyQt5 import QtWidgets
from PyQt5.QtWidgets import*
from PyQt5 import uic
import pics_rc
import time
import math
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from csv import writer
import sys
import threading

driver = webdriver.Chrome(executable_path = r"C:\Users\sheik\Downloads\chromedriver_win32\chromedriver.exe")
column_name = ["Name","Prices","Type","Company","Formula","Dose_Size","Quantity"]
df = pd.DataFrame(columns =column_name)
df.to_csv("medicine.csv",index = False)



class MyGUI(QDialog):
    global flag
    flag = True
    
    
    global label
    label = 0
    
    global page_no
    page_no = 1
    
    global i_check
    i_check = 0
    
    def __init__(self):
        super(MyGUI , self).__init__()
        uic.loadUi("Scraping_GUI.ui", self)
        self.start.clicked.connect(self.start_scraping)
        self.pause.clicked.connect(self.resume_scraping)
        self.stop .clicked.connect(self.stop_scraping)
       # self.ascending.toggled.connect(self.insertion_sort)
       # self.descending.toggled.connect(self.selection_sort)
        self.Namewise.clicked.connect(self.name_wise)
        self.Pricewise.clicked.connect(self.price_wise)
        self.Companywise.clicked.connect(self.company_wise)
        self.Formulawise.clicked.connect(self.formula_wise)
        self.Dosewise.clicked.connect(self.dose_wise)
        self.Categorywise.clicked.connect(self.category_wise)
        self.Quantitywise.clicked.connect(self.quantity_wise)
        self.progressBar.setRange(0,500)
        self.data_amount.setText("0 of 1053782")
        
        
        
      #  self.save_order.clicked.connect(self.SelectAlgo)
        
        
            
    def control_scrap(self):
        thread = threading.Thread(target = self.start_scraping,args = ())
        thread.start()
           
        
    def name_wise(self):
        df1 = pd.read_csv("Medicine.csv",usecols = [0])
        arr = df1.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [1])
        arr1 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [2])
        arr2 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [3])
        arr3 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [4])
        arr4 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [5])
        arr5 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [6])
        arr6 = df2.values.tolist()
        
        
        
        key1 = df1.values.tolist()
        arr = self.Select_Sorting(arr)
        
        key2 = []
        
        for i in range(len(arr)):
            for j in range(len(arr)):
                if arr[i][0]==key1[j][0]:
                    key2.append(j)
                    key1[j]="$$"
                    break 
            
        
        for i in range(len(arr)):
            
            self.datatable.setItem(i,0,QtWidgets.QTableWidgetItem(arr[i][0]))
            self.datatable.setItem(i,1,QtWidgets.QTableWidgetItem(arr1[key2[i]][0]))
            self.datatable.setItem(i,2,QtWidgets.QTableWidgetItem(arr2[key2[i]][0]))
            self.datatable.setItem(i,3,QtWidgets.QTableWidgetItem(arr3[key2[i]][0]))
            self.datatable.setItem(i,4,QtWidgets.QTableWidgetItem(arr4[key2[i]][0]))
            self.datatable.setItem(i,5,QtWidgets.QTableWidgetItem(arr5[key2[i]][0]))
            self.datatable.setItem(i,6,QtWidgets.QTableWidgetItem(arr6[key2[i]][0]))
            
        self.datatable.update()
        
    
    def price_wise(self):
        
        
        df2 = pd.read_csv("Medicine.csv", usecols = [1])
        arr = df2.values.tolist()
        
        df1 = pd.read_csv("Medicine.csv",usecols = [0])
        arr1 = df1.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [2])
        arr2 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [3])
        arr3 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [4])
        arr4 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [5])
        arr5 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [6])
        arr6 = df2.values.tolist()
        
        
        
        key1 = df2.values.tolist()
        arr = self.Select_Sorting(arr)
        
        key2 = []
        for i in range(len(arr)):
            for j in range(len(arr)):
                if arr[i][0]==key1[j][0]:
                    key2.append(j)
                    key1[j]="$$"
                    break         
        
        for i in range(len(arr)):
            
            self.datatable.setItem(i,1,QtWidgets.QTableWidgetItem(arr[i][0]))
            self.datatable.setItem(i,0,QtWidgets.QTableWidgetItem(arr1[key2[i]][0]))
            self.datatable.setItem(i,2,QtWidgets.QTableWidgetItem(arr2[key2[i]][0]))
            self.datatable.setItem(i,3,QtWidgets.QTableWidgetItem(arr3[key2[i]][0]))
            self.datatable.setItem(i,4,QtWidgets.QTableWidgetItem(arr4[key2[i]][0]))
            self.datatable.setItem(i,5,QtWidgets.QTableWidgetItem(arr5[key2[i]][0]))
            self.datatable.setItem(i,6,QtWidgets.QTableWidgetItem(arr6[key2[i]][0]))
            
        self.datatable.update()
        
    def company_wise(self):
        df1 = pd.read_csv("Medicine.csv",usecols = [2])
        arr = df1.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [1])
        arr1 = df2.values.tolist()
        
        df3 = pd.read_csv("Medicine.csv", usecols = [0])
        arr2 = df3.values.tolist()
        
        df4 = pd.read_csv("Medicine.csv", usecols = [3])
        arr3 = df4.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [4])
        arr4 = df2.values.tolist()
        
        df5 = pd.read_csv("Medicine.csv", usecols = [5])
        arr5 = df5.values.tolist()
        
        df6 = pd.read_csv("Medicine.csv", usecols = [6])
        arr6 = df6.values.tolist()
        
        
        
        key1 = df1.values.tolist()
        arr = self.Select_Sorting(arr)
        
        key2 = []
        
        for i in range(len(arr)):
            for j in range(len(arr)):
                if arr[i][0]==key1[j][0]:
                    key2.append(j)
                    key1[j]="$$"
                    break   

        
        for i in range(len(arr)):
            
            self.datatable.setItem(i,2,QtWidgets.QTableWidgetItem(arr[i][0]))
            self.datatable.setItem(i,1,QtWidgets.QTableWidgetItem(arr1[key2[i]][0]))
            self.datatable.setItem(i,0,QtWidgets.QTableWidgetItem(arr2[key2[i]][0]))
            self.datatable.setItem(i,3,QtWidgets.QTableWidgetItem(arr3[key2[i]][0]))
            self.datatable.setItem(i,4,QtWidgets.QTableWidgetItem(arr4[key2[i]][0]))
            self.datatable.setItem(i,5,QtWidgets.QTableWidgetItem(arr5[key2[i]][0]))
            self.datatable.setItem(i,6,QtWidgets.QTableWidgetItem(arr6[key2[i]][0]))
            
        self.datatable.update()
        
    
    def formula_wise(self):
        df1 = pd.read_csv("Medicine.csv",usecols = [3])
        arr = df1.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [1])
        arr1 = df2.values.tolist()
        
        df3 = pd.read_csv("Medicine.csv", usecols = [2])
        arr2 = df2.values.tolist()
        
        df3 = pd.read_csv("Medicine.csv", usecols = [0])
        arr3 = df3.values.tolist()
        
        df4 = pd.read_csv("Medicine.csv", usecols = [4])
        arr4 = df4.values.tolist()
        
        df5 = pd.read_csv("Medicine.csv", usecols = [5])
        arr5 = df5.values.tolist()
        
        df6 = pd.read_csv("Medicine.csv", usecols = [6])
        arr6 = df6.values.tolist()
        
        
        
        key1 = df1.values.tolist()
        arr = self.Select_Sorting(arr)
        
        key2 = []
        
        for i in range(len(arr)):
            for j in range(len(arr)):
                if arr[i][0]==key1[j][0]:
                    key2.append(j)
                    key1[j]="$$"
                    break 

        
        for i in range(len(arr)):
            
            self.datatable.setItem(i,3,QtWidgets.QTableWidgetItem(arr[i][0]))
            self.datatable.setItem(i,1,QtWidgets.QTableWidgetItem(arr1[key2[i]][0]))
            self.datatable.setItem(i,2,QtWidgets.QTableWidgetItem(arr2[key2[i]][0]))
            self.datatable.setItem(i,0,QtWidgets.QTableWidgetItem(arr3[key2[i]][0]))
            self.datatable.setItem(i,4,QtWidgets.QTableWidgetItem(arr4[key2[i]][0]))
            self.datatable.setItem(i,5,QtWidgets.QTableWidgetItem(arr5[key2[i]][0]))
            self.datatable.setItem(i,6,QtWidgets.QTableWidgetItem(arr6[key2[i]][0]))
            
        self.datatable.update()
        
    
    def dose_wise(self):
        df1 = pd.read_csv("Medicine.csv",usecols = [4])
        arr = df1.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [1])
        arr1 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [2])
        arr2 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [3])
        arr3 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [0])
        arr4 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [5])
        arr5 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [6])
        arr6 = df2.values.tolist()
        
        
        
        key1 = df1.values.tolist()
        arr = self.Select_Sorting(arr)
        
        key2 = []
        for i in range(len(arr)):
            for j in range(len(arr)):
                if arr[i][0]==key1[j][0]:
                    key2.append(j)
                    key1[j]="$$"
                    break  
             
            
        
        for i in range(len(arr)):
            
            self.datatable.setItem(i,4,QtWidgets.QTableWidgetItem(arr[i][0]))
            self.datatable.setItem(i,1,QtWidgets.QTableWidgetItem(arr1[key2[i]][0]))
            self.datatable.setItem(i,2,QtWidgets.QTableWidgetItem(arr2[key2[i]][0]))
            self.datatable.setItem(i,3,QtWidgets.QTableWidgetItem(arr3[key2[i]][0]))
            self.datatable.setItem(i,0,QtWidgets.QTableWidgetItem(arr4[key2[i]][0]))
            self.datatable.setItem(i,5,QtWidgets.QTableWidgetItem(arr5[key2[i]][0]))
            self.datatable.setItem(i,6,QtWidgets.QTableWidgetItem(arr6[key2[i]][0]))
            
        self.datatable.update()
        
    
    def category_wise(self):
        df1 = pd.read_csv("Medicine.csv",usecols = [5])
        arr = df1.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [1])
        arr1 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [2])
        arr2 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [3])
        arr3 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [4])
        arr4 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [0])
        arr5 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [6])
        arr6 = df2.values.tolist()
        
        
        
        key1 = df1.values.tolist()
        arr = self.Select_Sorting(arr)
        
        key2 = []
        for i in range(len(arr)):
             for j in range(len(arr)):
                if arr[i][0]==key1[j][0]:
                    key2.append(j)
                    key1[j]="$$"
                    break 
            
        
        for i in range(len(arr)):
            
            self.datatable.setItem(i,5,QtWidgets.QTableWidgetItem(arr[i][0]))
            self.datatable.setItem(i,1,QtWidgets.QTableWidgetItem(arr1[key2[i]][0]))
            self.datatable.setItem(i,2,QtWidgets.QTableWidgetItem(arr2[key2[i]][0]))
            self.datatable.setItem(i,3,QtWidgets.QTableWidgetItem(arr3[key2[i]][0]))
            self.datatable.setItem(i,4,QtWidgets.QTableWidgetItem(arr4[key2[i]][0]))
            self.datatable.setItem(i,0,QtWidgets.QTableWidgetItem(arr5[key2[i]][0]))
            self.datatable.setItem(i,6,QtWidgets.QTableWidgetItem(arr6[key2[i]][0]))
            
        self.datatable.update()
        
    def quantity_wise(self):
        df1 = pd.read_csv("Medicine.csv",usecols = [6])
        arr = df1.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [1])
        arr1 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [2])
        arr2 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [3])
        arr3 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [4])
        arr4 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [5])
        arr5 = df2.values.tolist()
        
        df2 = pd.read_csv("Medicine.csv", usecols = [0])
        arr6 = df2.values.tolist()
        
        
        
        key1 = df1.values.tolist()
        arr = self.Select_Sorting(arr)
        
        key2 = []
    
        for i in range(len(arr)):
             for j in range(len(arr)):
                if arr[i][0]==key1[j][0]:
                    key2.append(j)
                    key1[j]="$$"
                    break 
            
        
        for i in range(len(arr)):
            
            self.datatable.setItem(i,6,QtWidgets.QTableWidgetItem(arr[i][0]))
            self.datatable.setItem(i,1,QtWidgets.QTableWidgetItem(arr1[key2[i]][0]))
            self.datatable.setItem(i,2,QtWidgets.QTableWidgetItem(arr2[key2[i]][0]))
            self.datatable.setItem(i,3,QtWidgets.QTableWidgetItem(arr3[key2[i]][0]))
            self.datatable.setItem(i,4,QtWidgets.QTableWidgetItem(arr4[key2[i]][0]))
            self.datatable.setItem(i,5,QtWidgets.QTableWidgetItem(arr5[key2[i]][0]))
            self.datatable.setItem(i,0,QtWidgets.QTableWidgetItem(arr6[key2[i]][0]))
            
        self.datatable.update()
        
        
        
        
         
        
    
    
    def Select_Sorting(self,arr):
        if self.select_algorithm.currentText()=="Insertion Sort":
            return self.insertion_sort(arr)
        elif self.select_algorithm.currentText()=="Selection Sort":
            return self.selection_sort(arr)
        elif self.select_algorithm.currentText()=="Bubble Sort":
            return self.bubble_sort(arr)
        elif self.select_algorithm.currentText()=="Merge Sort":
            size = len(arr)
            self.MergeSort(arr,0,size-1)
            return arr
        elif self.select_algorithm.currentText()=="Quick Sort":
            size = len(arr)
            self.quickSort(arr,0,size-1)
            return arr
        elif self.select_algorithm.currentText()=="Counting Sort":
            self.countingSort(arr)
            return arr
            
        
    
    def start_scraping(self):
        alpha = []
        count = 97  

        for i in range(26):
            alpha.append(chr(count))
            count = count+1
        global flag
        global label
        global page_no
        
        flag = True
        
        
        for i in range(26):
            driver.get("https://www.1mg.com/drugs-all-medicines?label="+alpha[i])
            
            content = driver.page_source
            soup = BeautifulSoup(content)
            page = soup.find_all('a',{'class':'button-text link-page'})
            check = int(page[len(page)-1].text)  
            
            for x in range(1,check+1): 
                
                self.getmedicine(x,alpha[i])
                self.datatable.update()
                r_check = self.datatable.rowCount()
                self.progressBar.setValue(r_check)
                self.data_amount.setText(str(r_check)+" of 242500")
                self.data_amount.update()
                self.progressBar.update()
                
                
                if flag ==False:
                    page_no = x
                    break
            
            if flag ==False:
                label = i
                break
                
    
    def resume_scraping(self):
        alpha = []
        count = 97  
        global flag
        global label
        global page_no
        
        if flag ==False:
            flag = True
        else:
            flag = True
            
        

        for i in range(1,26):
            alpha.append(chr(count))
            count = count+1
        
        
        if flag== True:
            
            for i in range(label,26):
                driver.get("https://www.1mg.com/drugs-all-medicines?label="+alpha[i])
                content = driver.page_source
                soup = BeautifulSoup(content)
                page = soup.find_all('a',{'class':'button-text link-page'})
                check = int(page[len(page)-1].text)  
                for x in range(page_no,check+1): 
                    self.getmedicine(x,alpha[i])
                    self.datatable.update()
                    r_check = self.datatable.rowCount()
                    self.progressBar.setValue(r_check)
                    self.data_amount.setText(str(r_check)+" of 242500")
                    self.data_amount.update()
                    self.progressBar.update()


                    if flag ==False:
                        page_no = x 

                if flag ==False:
                    label = i
                    break
            
        
    
    
    def stop_scraping(self):
        global label
        label = 0
        
        global flag
        flag = False
        
        global page_no
        page_no = 1
        
        
        
    def quickSort(self,arr, low, high):
        if (low < high):
            pi = self.partition(arr, low, high);
            self.quickSort(arr, low, pi - 1)
            self.quickSort(arr, pi + 1, high)

    def partition(self,arr,low,high):
    
        pivot = arr[high]
        i =low-1
        for j in range(low,high):
            if arr[j]<pivot:
                i = i+1
                temp = arr[i]
                arr[i] = arr[j]
                arr[j] = temp

        temp = arr[i+1]
        arr[i+1] = arr[high]
        arr[high] = temp
        return (i+1)
    
    
    def insertion_sort(self,arr):
        
        new_arr = []
        for j in range(1,len(arr)):
            key = arr[j]
            i = j-1
            while i>=0 and arr[i]>key:
                arr[i+1] = arr[i]
                i = i-1
            arr[i+1] = key
        return arr
        
    def selection_sort(self,arr):
         
        
        for i in range(len(arr)):
                   
            min_idx = i
            
            
            for j in range(i+1,len(arr)):
                check1 = self.slicer(arr[min_idx][0])
                check2 = self.slicer(arr[j][0])
               
                if check2<check1:
                    min_idx = j
                    
             # put min at the correct position
            (arr[i], arr[min_idx]) = (arr[min_idx], arr[i])
            
        return arr
    
    def bubble_sort(self,arr):
        n = len(arr)

        for i in range(n-1):

            for j in range(0, n-i-1):

                if arr[j] > arr[j + 1] :
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]
                                  
            
        return arr
    
    def Merge(self,arr,l,mid,r):
        n1 = mid-l+1
        n2 = r-mid

        arr1 = []
        arr2 = []
        for n in range(n1):
            arr1.append(arr[l+n])
        for n in range(n2):
            arr2.append(arr[mid+1+n])
        i=0
        j=0
        k=l
        while i<n1 and j<n2:
            if arr1[i]<arr2[j]:
                arr[k] = arr1[i]
                i = i+1
                k = k+1
            else :
                arr[k] = arr2[j]
                j = j+1
                k = k+1
        while i<n1:
            arr[k] = arr1[i]
            i = i+1
            k = k+1
        while j<n2:
            arr[k] = arr2[j]
            j = j+1
            k = k+1
            
    
    def MergeSort(self,arr,l,r):
         if l<r:
            mid = math.floor((l+r)/2)
            self. MergeSort(arr,l,mid)
            self.MergeSort(arr,mid+1,r)
            self.Merge(arr,l,mid,r)
            
            
    def countingSort(self,arr):
        size = len(arr)
        output = [0] * size

        new_arr = []
        for i in range(len(arr)):
            new_arr.append(arr[i][0]) 

        new_arr2 = []
        for i in range(len(new_arr)):
            word_check = new_arr[i]
            new_arr2.append(ord(word_check[i]))



        # count array initialization
        count = [0] * 10

        # storing the count of each element 
        for m in range(0, size):
            count[new_arr2[m]] += 1

        # storing the cumulative count
        for m in range(1, 10):
            count[m] += count[m - 1]

        # place the elements in output array after finding the index of each element of original array in count array
        m = size - 1
        while m >= 0:
            output[count[new_arr2[m]] - 1] = new_arr2[m]
            count[new_arr2[m]] -= 1
            m -= 1
            
        key2 = []
        for i in range(len(arr)):
            for j in range(len(arr)):
                if output[i]==new_arr2[j]:
                    key2.append(j)
                    break
        

        for m in range(0, size):
            arr[m] = arr[key2[m]][0]
        return
   
    
        
    def slicer(self,arr):
        price2 =""
        for b in arr:
            if b.isdigit():
                price2 = price2+b

        check2 = float(price2)
        return check2
        
        
    def getmedicine(self,pages,labels):
        
        global i_check
        driver.get("https://www.1mg.com/drugs-all-medicines?page="+str(pages)+"&label="+labels)
        content = driver.page_source
        soup = BeautifulSoup(content,'html.parser')
        medList=soup.findAll('div',attrs={'class' : 'style__width-100p___2woP5 style__flex-row___m8FHw'})

        row = self.datatable.rowCount()
        for a in medList:
            name=(a.find('div', attrs={'class':'style__font-bold___1k9Dl style__font-14px___YZZrf style__flex-row___2AKyf style__space-between___2mbvn style__padding-bottom-5px___2NrDR'}).findChildren()[0])
            price=(a.find('div', attrs={'class':'style__font-bold___1k9Dl style__font-14px___YZZrf style__flex-row___2AKyf style__space-between___2mbvn style__padding-bottom-5px___2NrDR'}).findChildren()[1])
            Type=a.find('div', attrs={'class':'style__font-12px___2ru_e style__product-content___5PFBW style__display-inline-block___2y7gd'})
            Company=(a.find('div',attrs={'class':'style__flex-column___1zNVy style__font-12px___2ru_e'}).findChildren()[1])
            Formula=a.find('div',attrs={'class':'style__font-12px___2ru_e style__product-content___5PFBW style__display-inline-block___2y7gd'})
            DoseSize=a.find('div',attrs={'class':'style__font-12px___2ru_e style__product-content___5PFBW style__display-inline-block___2y7gd'})
            Quantity=(a.find('div',attrs={'class':'style__flex-column___1zNVy style__font-12px___2ru_e'}).findChildren()[0])

            list_data = [name.text,price.text,Company.text,Type.text,Formula.text,DoseSize.text,Quantity.text]

            self.datatable.insertRow(row)

            self.datatable.setItem(i_check,0,QtWidgets.QTableWidgetItem(name.text))
            self.datatable.setItem(i_check,1,QtWidgets.QTableWidgetItem(price.text))
            self.datatable.setItem(i_check,2,QtWidgets.QTableWidgetItem(Company.text))
            self.datatable.setItem(i_check,3,QtWidgets.QTableWidgetItem(Type.text))
            self.datatable.setItem(i_check,4,QtWidgets.QTableWidgetItem(Formula.text))
            self.datatable.setItem(i_check,5,QtWidgets.QTableWidgetItem(DoseSize.text))
            self.datatable.setItem(i_check,6,QtWidgets.QTableWidgetItem(Quantity.text))
            
            row = row+1
            i_check = i_check+1

            with open('medicine.csv','a',newline="",encoding='utf-8') as f:
                writer_object = writer(f)

                writer_object.writerow(list_data)
                f.close()

        return
    
    
    
        
def main():
    app = QApplication([])
    window = MyGUI()
    
    window.show()
    app.exec_()
    

    

if __name__ == '__main__':
    main()
```
